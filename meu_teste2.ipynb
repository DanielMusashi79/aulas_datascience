```python
import pandas as pd
data = {'Nome': ["Manji", "Guts"], 'Salário': ['R$ 2.800', 'R$ 2.100']}
df = pd.DataFrame(data)
df # Apenas digitar 'df' no final mostra a tabela formatada
```

```python
df['Salário_Num'] = df['Salário'].str.replace('R$ ', '', regex=False).str.replace('.', '', regex=False).astype(float)
print(f"Média: {df['Salário_Num'].mean()}")
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Nome'], df['Salário_Num'], color='green')
plt.title('Salários dos Protagonistas')
plt.show()
```
