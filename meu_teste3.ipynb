```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Nome': ["Manji", "Magatsu", "Musashi", "Guts"],
    'Salário': ['R$ 2.800', 'R$ 2.300', 'R$ 6.550', 'R$ 2.100']
}
df = pd.DataFrame(data)
```

```python
# Limpeza rápida
df['Salário_Num'] = df['Salário'].str.replace('R$ ', '', regex=False).str.replace('.', '', regex=False).astype(float)
df
```

```python
plt.bar(df['Nome'], df['Salário_Num'], color='green')
plt.title('Salários dos Protagonistas')
plt.show()
```
